```python
Sum = df2['count'].sum()
print(Sum)
```

# 3) What is the most common Genre for Vin Diesel & Emma Watson movies?

```python
df3 = movie_df[['cast','genres']]
df3
```

```python
sub = 'Vin Diesel'
df3["Indexes"] = df3["cast"].str.find(sub)
df3
```

```python
df3vd = df3[df3['Indexes'] == 0]
df3vd
```

```python
df3vd.count()
```

```python
g_and_c = {}
for i in range(df3vd.shape[0]):
    genres = str(df3vd['genres'].values[i]).split('|')
    for j in genres:
        try:
            count = g_and_c[j]
            g_and_c[j] = count + 1
        except:
            g_and_c[j] = 1
        
g_and_c
```

### From this we can conclude that the most common genre for Vin Diesel is action with 15 entries

### Now for Emma Watson,
